```python
import pandas as pd

# save filepath to variable for easier access
melbourne_file_path = './melb_data.csv'

# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path)

# print a summary of the data in Melbourne data
melbourne_data.describe()


```

```python
# columnas del dataset
melbourne_data.columns
```

```python
# dropna drops missing values (think of na as "not available")
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
# en y quiero un DF con una sola columna.
# es nuestra "prediction target"
y = melbourne_data.Price
```

```python
# las features que voy a usar para predecir  el precio luego .
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

# en X voy a guardar la data para predecir
X = melbourne_data[melbourne_features]

# x.describe()
```

```python
# sklearn es la lib mas pop para modelas datos tipicamente guardados en DF.

from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)

```

```python
# pido predicciones de los primeros cinco qye ya tiene "aprendidos".

print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
# listo los primeros cinco de mi dataset inicial para comprarar
melbourne_data.head()
```

```python
# hago una copia de data de solo cinco registros.
# luego cambio rooms de 3 a 5

xx = melbourne_data.head()

xx = xx[melbourne_features]

xx['Rooms'] = xx['Rooms'].replace(3,5)


# finalmente hago nueva predicción sobre datos nuevos

print("The predictions are")
print(melbourne_model.predict(xx))

# tengo predicciones distintas, yeah.
```
